```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/home/kou552/crypto/archive/ETH_1H.csv")
df.head()
```

```python
# Unix Timestampdatetime
df['Timestamp'] = pd.to_datetime(df['Date'])

# 
fig, axs = plt.subplots(6, 1, figsize=(10, 15), sharex=True)

# 
axs[0].plot(df['Timestamp'], df['Open'])
axs[0].set_ylabel('Open')
axs[1].plot(df['Timestamp'], df['High'])
axs[1].set_ylabel('High')
axs[2].plot(df['Timestamp'], df['Low'])
axs[2].set_ylabel('Low')
axs[3].plot(df['Timestamp'], df['Close'])
axs[3].set_ylabel('Close')
axs[4].plot(df['Timestamp'], df['Volume'])
axs[4].set_ylabel('Volume')

# 
plt.xticks(rotation=45)

# 
plt.tight_layout()

# 
plt.show()
```
